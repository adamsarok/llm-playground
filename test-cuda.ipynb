```python
import torch

print(torch.__version__)
print(torch.version.cuda)

# Check if CUDA is available
print(f"CUDA available: {torch.cuda.is_available()}")

# Check number of GPUs
print(f"Number of GPUs: {torch.cuda.device_count()}")

# Get current device name
if torch.cuda.is_available():
    print(f"Current device: {torch.cuda.get_device_name(0)}")
```

```python
# Check default device for tensors
x = torch.rand(3, 3)
print(f"Tensor device: {x.device}")  # Should show 'cpu' or 'cuda:0'

# Create tensor on GPU explicitly
if torch.cuda.is_available():
    x_gpu = torch.rand(3, 3).cuda()
    print(f"GPU tensor device: {x_gpu.device}")
```

```python
capability = torch.cuda.get_device_capability()
use_bf16 = capability[0] >= 8 # bfloat16 support
print(f"bfloat16 support: {use_bf16}")
```
